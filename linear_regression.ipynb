```python
#read acsv file in to a dataframe
import pandas as pd
df=pd.read_csv('Book1.csv')

df=df/df.max()
print(df)
```

```python
import matplotlib

import matplotlib.pyplot as plt
#creates a line plot
plt.plot(df['AREA'],df['PRICE'])
```

```python
#creates a scatter plot
plt.scatter(df['AREA'],df['PRICE'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Creating an instance of the model
model = LinearRegression()
#train the model
model.fit(df[['AREA']], df[['PRICE']])
```

```python
plt.scatter(df['AREA'],df['PRICE'])
plt.plot(df['AREA'],model.predict(df[['AREA']]))
```

```python
test_df=pd.read_csv('Book2.csv')
print(model.predict(test_df[['AREA']]))
```

```python
def gradient_descent(df):
  m=c=1
  learning_rate=0.01
  iterations=1000
  x_values=[]
  y_values=[]
  for i in range (iterations):
    cost=0
    sum_m=0
    sum_c=0
    n=len(df)
    for index in df.index:
      x=df['AREA'][index]
      y=df['PRICE'][index]
      y_predicted=m*x+c
      cost+=(1/n)*(y-y_predicted)**2
      sum_m+=x*(y-y_predicted)
      sum_c+=(y-y_predicted)
    nd=(-2/n)*sum_m
    nc=(-2/n)*sum_c
    m=m-learning_rate*nd
    c=c-learning_rate*nc
    x_values.append(i)
    y_values.append(cost)

  #plt.plot(x_values,y_values)
  #plt.xlabel('no_of_iterations')
  #plt.ylabel('MSE')
  #plt.show()
  return m,c

m,c=gradient_descent(df)

def predict(x,m,c):
  x=x/4000
  y=(m*x)+c
  print(y*72500)


predict(5100,m,c)
predict(5400,m,c)


```
